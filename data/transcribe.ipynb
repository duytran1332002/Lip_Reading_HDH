```python
from speech_to_text import SpeechToText
import os
```

```python
def transcribe(audios_path, transcripts_path, alignments_path):
    '''
        Transcribe audio files and save as text files
        
        Parameters:
            audios_path: str
                Path to audio folder
            save_path: str
                Path to save result
    '''    
    assert os.path.exists(audios_path)
    if not os.path.exists(transcripts_path):
        os.makedirs(transcripts_path)
    if not os.path.exists(alignments_path):
        os.makedirs(alignments_path)
    transcriber = SpeechToText()
    for audio in os.listdir(audios_path):
        audio_path = os.path.join(audios_path, audio)
        transcript_path = os.path.join(transcripts_path, audio.replace('.wav', '.txt'))
        alignment_path = os.path.join(alignments_path, audio.replace('.wav', '.csv'))
        transcriber.save_result_to_file(audio_path=audio_path, 
                                        transcript_path=transcript_path,
                                        alignment_path=alignment_path)
```

```python
# Transcribe and align
audios_path = r'..\..\data\audios'
transcripts_path = r'..\..\data\transcripts'
alignments_path = r'..\..\data\alignments'
transcribe(audios_path=audios_path,
           transcripts_path=transcripts_path,
           alignments_path=alignments_path)
```

```python
def second_to_timecode(x: float) -> str:
    hour, x = divmod(x, 3600)
    minute, x = divmod(x, 60)
    second, x = divmod(x, 1)
    millisecond = int(x * 1000.)
    return '%.2d:%.2d:%.2d,%.3d' % (hour, minute, second, millisecond)

def convert_to_srt(alignement_path):
    df = pd.read_csv(alignement_path, names=['start', 'end', 'word'])
    with open('test.srt', 'w', encoding='utf-8') as f:
        for i in range(len(df)):
            f.write(f'{i+1}\n')
            f.write(f'{second_to_timecode(df["start"].values[i])} --> {second_to_timecode(df["end"].values[i])}\n')
            f.write(f'{df["word"].values[i]}\n\n')
alignment_path = r'..\..\test_data\aligned_transcripts\20230105.csv'
convert_to_srt(alignment_path)
```
